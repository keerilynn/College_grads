# College graduate outcomes for 2010-2012

This notebook will take a look at data taken by the American Community Survey from students that graduated college in 2010-2012. FiveTHirtyEight cleaned the dataset and released it on their Githib repo.

The columns are all cleaned, and here they are broken down: 

- Rank - Rank by median earnings (the dataset is ordered by this column). 
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.


```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline 
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
```

```python
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
raw_data_count = recent_grads.count()
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.count()
print(cleaned_data_count)
```

# Exploring the data with scatter plots

First, we'll explore with sample size and median

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
```

Next, we will look at Samle_size and Unemployment_rate.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate')
```

Here comes Full_time and Median. 

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
```

Next, we'll look at ShareWomen and Unemployment_rate.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate')
```

Next, we will look at unemployment rate in women and men. 

```python
fig = plt.figure()

ax1= recent_grads.plot(x='Women', y='Unemployment_rate', kind='scatter', title='Women vs. Unemployment_rate')
ax2= recent_grads.plot(x='Men', y='Unemployment_rate', kind='scatter', title='Men vs. Unemployment_rate')
ax2.set_xlim(-50000, 350000)
```

Now, we'll do women and men versus the median. 

```python
fig=plt.figure()

ax1 = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
ax2 = recent_grads.plot(x="Men", y='Median', kind='scatter', title='Men vs. Median')
ax2.set_xlim(-50000, 350000)
```

Women seem to have a slightly higher unemployment rate.

# Exploring with scatter matrix plots

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
plt.suptitle('Sample_size vs. Median')
```

# Comparison with bar plots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen').set_title('First 10 Major vs. ShareWomen')

recent_grads[-10:].plot.bar(x='Major', y='ShareWomen').set_title('Last 10 Major vs. ShareWomen')
```

A higher share of women finished with degrees in the last ten majors than the first ten majors. The first ten were more STEM related, and the last ten were not as much.
